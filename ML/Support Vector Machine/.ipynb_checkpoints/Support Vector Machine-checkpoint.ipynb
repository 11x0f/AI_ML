## FEATURE ENGINEERING

```python
Male = pd.get_dummies(data["sex"], drop_first=True)
data = pd.concat([data, Male], axis = 1)
data.head()
```

```python
Smoker = pd.get_dummies(data["smoker"], drop_first=True)
data = pd.concat([data,Smoker], axis =1)
data.head()
```

```python
data.rename(columns={"yes":"smokers"}, inplace=True)
```

###### remove low corrr colms

```python
data.drop(columns=["sex"],inplace =True)
data.head()
```

# Splitting up the data

```python
x = data.drop("charges", axis=1)
y = data["charges"]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
x_test.shape
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

```python
scaled_x_train
```

# support vector regression model

```python
from sklearn.svm import SVR

regressor = SVR()
regressor.fit(scaled_x_train, y_train)
```

```python
predict = regressor.predict(scaled_x_test)
```

```python
predict.shape
```

```python
scaled_x_test.shape
```

```python
predict[:10]
```

```python
y_test[:10]
```

```python
plt.plot(x_test, y_test, "*", color = "green")
plt.plot(x_test, predict, ".", color="blue")
plt.show()
```

```python
plt.scatter(scaled_x_test, y_test, color="red")
plt.plot(scaled_x_test, predict, color="blue")
plt.xlabel("input")
plt.ylabel("predictions")
plt.show()
```
