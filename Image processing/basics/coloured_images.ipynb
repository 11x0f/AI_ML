```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## creating a red image

```python
red = np.array([[ [255,0,0], [255,0,0], [255,0,0]],
                [ [255,0,0], [255,0,0], [255,0,0]],
                [ [255,0,0], [255,0,0], [255,0,0]] ])
plt.imshow(red)
cv.imwrite("red_image.png", red)
```

## creating red blue green pattern

```python
pattern = np.array([ [[255,0,0], [255,0,0], [255,0,0]],
                     [[0,255,0], [0,255,0], [0,255,0]],
                     [[0,0,255], [0,0,255], [0,0,255]] 
])
plt.imshow(pattern)
```

## creating different red 

```python
shades_red = np.array([[ [30,0,0], [60,0,0], [90,0,0]],
                [ [120,0,0], [150,0,0], [180,0,0]],
                [ [210,0,0], [240,0,0], [255,0,0]] ])

plt.imshow(shades_red)
```

## creating different shades of all 3 colors

```python
img =[]

for i in range(256):
    t = []
    for j in range(256*3):
        t.append([i,0,0])
    img.append(t)
for i in range(256,0,-1):
    t = []
    for j in range(256*3):
        t.append([i,0,0])
    img.append(t)

for i in range(256):
    t = []
    for j in range(256*3):
        t.append([0,i,0])
    img.append(t)
for i in range(256,0,-1):
    t = []
    for j in range(256*3):
        t.append([0,i,0])
    img.append(t)

for i in range(256):
    t = []
    for j in range(256*3):
        t.append([0,0,i])
    img.append(t)
for i in range(256,0,-1):
    t = []
    for j in range(256*3):
        t.append([0,0,i])
    img.append(t)
img = np.array(img)
plt.imshow(img)
cv.imwrite("rgb_transition.png",img)
```
