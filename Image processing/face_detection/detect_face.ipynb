```python
import cv2 as cv
import numpy as np
```

```python
classifier = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
cam = cv.VideoCapture(0)
while True:
    _, img = cam.read()
    img = cv.flip(img, 1)
    faces = classifier.detectMultiScale(img, 1.1, 8)
    
    print(faces)
    print("---------------------------")

    for (x, y, w, h) in faces :
        cv.rectangle(img, (x,y), (x+w, y+h) , (255,255,255), 5)
    cv.imshow("Frame ", img)
    if cv.waitKey(1) ==27:
        cam.release()
        cv.destroyAllWindows()
        break
```
