```python
import cv2 as cv
import numpy as np
```

```python
classifier = cv.CascadeClassifier("haarcascade_frontalface_default.xml") # for front face
#classifier2 = cv.CascadeClassifier("haarcascade_profileface.xml")   # for side face
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img, 1)
    front_faces = classifier.detectMultiScale(img, 1.4, 5)
    #side_faces = classifier2.detectMultiScale(img, 1.4, 5)
    #print (front_faces, side_faces)
    print(front_faces)
    for (x,y,w,h) in front_faces :
        cv.rectangle(img, (x,y), (x+w, y+h), (255,255,255), 5)
        front_face = img[y:y+h, x:x+w, :]
        cv.imshow("front face", front_face)
        front_face = cv.blur(front_face, (16,16))
        img[y:y+h, x:x+w, :] = front_face

    # for (x_, y_, w_, h_) in side_faces:
    #     cv.rectangle(img, (x,y), (x+w, y+h), (0, 0, 0), 5)
    #     side_face = img[y_:y_+h_, x_:x_+w_, :]
    #     cv.imshow("side face", side_face)
    
    
    cv.imshow("Camera View", img)
    
    

    if cv.waitKey(1) == 27 :
        cam.release()
        cv.destroyAllWindows()
        break
```
